```python
import librosa
import os
import numpy as np
import pandas as pd
```

```python
path="Training Set"
X=np.zeros(4440)
members=[]
for i in os.listdir(path):
    signal, sr=librosa.load(os.path.join(path,i))
    mfcc= librosa.feature.mfcc(signal,sr,n_mfcc=40)
    mfcc=mfcc.reshape(-1)
    mfcc=mfcc[:4440]
    X=np.vstack([X,mfcc])
    members.append(i.split(" ")[0])
    
```

```python
X=X[1:]
```

```python
X=pd.DataFrame(X)
X["members"]=members
X.head()
```

```python
X=X.drop(columns=["members"],axis=1)
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
pca= PCA(n_components=30)
pca.fit(X)
```

```python
import pickle
pickle.dump(pca, open("pca.pkl","wb"))
```

```python
from sklearn.feature_selection import SelectKBest,f_classif
selector= SelectKBest(f_classif,k=5)
selector.fit(X.drop("members",axis=1),X["members"])
```
